```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.vstack((
    -1 + np.random.rand(10,2),
    1 + np.random.rand(10,2)
))
y = np.zeros((20, 1))
y[:10] = 1

mu1 = np.mean(x[:10], axis=0)
mu2 = np.mean(x[10:], axis=0)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x[:10, 0], x[:10, 1], c='b', marker='x')
ax.scatter(mu1[0], mu1[1], c='b')
ax.scatter(x[10:, 0], x[10:, 1], c='r', marker='x')
ax.scatter(mu2[0], mu2[1], c='r')
ax.set_aspect('equal')
plt.show()
```

```python
xlist = np.linspace(-2.0, 2.0, 100)
ylist = np.linspace(-2.0, 2.0, 100)
X, Y = np.meshgrid(xlist, ylist)

def distance(x, y):
    return np.linalg.norm(x**2-y**2)
```

```python
Z = np.zeros((100, 100))
for i in range(100):
    for j in range(100):
        if distance(np.array([X[i,j],Y[i,j]]), mu1) > distance(np.array([X[i,j],Y[i,j]]), mu2):
            Z[i,j] = 1
```

```python
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) 
ax.scatter(x[:10, 0], x[:10, 1], c='b', marker='x')
ax.scatter(mu1[0], mu1[1], c='b')
ax.scatter(x[10:, 0], x[10:, 1], c='r', marker='x')
ax.scatter(mu2[0], mu2[1], c='r')
plt.show()
```
